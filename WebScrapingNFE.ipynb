```
import requests #Library standard making HTTP requests in Python
from bs4 import BeautifulSoup #library for colect data in site (webscrap)
```

```
#Importing the URL of page and reading the content 

url = ('http://app.sefaz.es.gov.br/ConsultaNFCe/qrcode.aspx?p=32200701928075006211650040001169441867136719|2|1|1|691ebce814f3124630a5562a58982707d67cceb6')

response = requests.get(url)
html = response.content #to list the page and content of the site
scraped = BeautifulSoup(html, 'html.parser') #content for html 
print(scraped.prettify()) 
```

```
#Reading the place of buying
empresa=scraped.find(id="conteudo")
local_compra=empresa.find('div', class_="txtTopo")
print("Local da compra:", local_compra.text.strip(), end='\n'*2)


#Reading the products
tabela = scraped.find(id='tabResult')
produtos=(tabela.find_all('tr'))
for produto in produtos:
  item = produto.find('span', class_='txtTit')
  qtd  = produto.find('span', class_='Rqtd',)
  unid = produto.find('span', class_='RUN')
  vl_unit = produto.find('span',class_='RvlUnit')
  vl_total_produto = produto.find('span',class_='valor')

  print("Produto comprado:",item.text.strip())
  print(qtd.text.strip())
  print(unid.text.strip())
  print(vl_unit.text.strip())
  print("Total Produto:" ,vl_total_produto.text.strip(),end = '\n'*2)
```
